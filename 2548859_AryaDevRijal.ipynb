6.Adding HDI Category Column:

• Create a new column HDI Category that classifies each country into one of the four official Human Development Index groups. The classification should be based on the HDI value for the year 2022. Use the following categories and thresholds defined by the United Nations Development Programme (UNDP):

HDI Category HDI Range (hdi)

Low < 0.550

Medium 0.550 – 0.699

High 0.700 – 0.799

Very ≥ 0.800

After creating this new column:

• verify that all countries are classified correctly,

• ensure the updated dataframe includes the new category column.

• Save the final dataframe as HDI category added.csv and include this file in your final
submission.

```python
# Function to classify HDI
def classify_hdi(hdi_value):
    if hdi_value < 0.550:
        return 'Low'
    elif 0.550 <= hdi_value <= 0.699:
        return 'Medium'
    elif 0.700 <= hdi_value <= 0.799:
        return 'High'
    elif hdi_value >= 0.800:
        return 'Very High'
    else:
        return 'Unknown'
```

```python
# Create HDI Category column
hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(classify_hdi)
```

```python
# Check how many countries fall in each category
category_counts = hdi_2022_df['HDI_Category'].value_counts()
print("Number of countries in each HDI Category:\n", category_counts)

# Preview the updated DataFrame
print(hdi_2022_df[['country', 'hdi', 'HDI_Category']].head(10))
```

3.Outlier Detection:

• Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.

• Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.

• Discuss why the identified countries stand out as outliers.

```python
south_asia_2022 = south_asia_df[south_asia_df["year"] == 2022].copy()
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return (df[column] < lower_bound) | (df[column] > upper_bound)
```

```python
south_asia_2022["hdi_outlier"] = detect_outliers_iqr(
    south_asia_2022, "hdi"
)

south_asia_2022["gni_outlier"] = detect_outliers_iqr(
    south_asia_2022, "gross_inc_percap"
)

# Combine both
south_asia_2022["outlier"] = (
    south_asia_2022["hdi_outlier"] |
    south_asia_2022["gni_outlier"]
)
```

```python
south_asia_2022.loc[
    south_asia_2022["outlier"],
    ["country", "hdi", "gross_inc_percap"]
]
```

```python
import matplotlib.pyplot as plt

# Normal points
normal = south_asia_2022[~south_asia_2022["outlier"]]
outliers = south_asia_2022[south_asia_2022["outlier"]]

plt.figure(figsize=(8, 6))

plt.scatter(
    normal["gross_inc_percap"],
    normal["hdi"],
    label="Normal Countries"
)

plt.scatter(
    outliers["gross_inc_percap"],
    outliers["hdi"],
    label="Outliers"
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita (South Asia, 2022)")
plt.legend()
plt.show()
```

• Create a bar chart comparing these top and bottom performers across the two regions.

```python
import pandas as pd
import matplotlib.pyplot as plt

south_asia_df = pd.read_csv("HDI SouthAsia 2020 2022.csv")
middle_east_df = pd.read_csv("HDI MiddleEast 2020 2022.csv")
```

```python
# South Asia
sa_avg_hdi = south_asia_df.groupby("country")["hdi"].mean().reset_index()

# Middle East
me_avg_hdi = middle_east_df.groupby("country")["hdi"].mean().reset_index()
```

```python
# South Asia
sa_top3 = sa_avg_hdi.nlargest(3, "hdi")
sa_bottom3 = sa_avg_hdi.nsmallest(3, "hdi")

# Middle East
me_top3 = me_avg_hdi.nlargest(3, "hdi")
me_bottom3 = me_avg_hdi.nsmallest(3, "hdi")
```

```python
top_bottom_df = pd.concat([
    sa_top3.assign(Region="South Asia"),
    sa_bottom3.assign(Region="South Asia"),
    me_top3.assign(Region="Middle East"),
    me_bottom3.assign(Region="Middle East")
])

# For better readability
top_bottom_df = top_bottom_df.sort_values(
    by=["Region", "hdi"], ascending=[True, False]
)
```

```python
plt.figure(figsize=(10, 6))
colors = {"South Asia": "skyblue", "Middle East": "orange"}

plt.bar(
    top_bottom_df["country"],
    top_bottom_df["hdi"],
    color=[colors[r] for r in top_bottom_df["Region"]]
)

plt.xlabel("Country")
plt.ylabel("Average HDI (2020–2022)")
plt.title("Top and Bottom HDI Performers in South Asia vs Middle East")
plt.xticks(rotation=45, ha="right")
plt.legend(colors.keys())
plt.show()
```

The bar chart shows the top and bottom performers in South Asia and the Middle East based on average HDI from 2020–2022. In South Asia, Sri Lanka, Maldives, and India are the top performers, while Afghanistan, Pakistan, and Nepal are the lowest. In the Middle East, Israel, United Arab Emirates, and Qatar have the highest HDI, whereas Yemen, Syria, and Palestine have the lowest. The chart highlights stark disparities in human development both within and between the two regions.

4.Metric Comparisons:

• Compare the following metrics across regions using grouped bar charts:

– Gender Development Index {"gender development"}

– Life Expectancy Index {"life expectancy"}

– GNI per Capita Index {"gross inc percap"}

• Identify which metric shows the greatest disparity between regions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

south_asia_df = pd.read_csv("HDI SouthAsia 2020 2022.csv")
middle_east_df = pd.read_csv("HDI MiddleEast 2020 2022.csv")
```

```python
metrics = ["gender_development", "life_expectancy", "gross_inc_percap"]

# Compute regional means
south_asia_mean = south_asia_df[metrics].mean()
middle_east_mean = middle_east_df[metrics].mean()

comparison_df = pd.DataFrame({
    "South Asia": south_asia_mean,
    "Middle East": middle_east_mean
}).reset_index().rename(columns={"index": "Metric"})

comparison_df
```

```python
comparison_df_norm = comparison_df.copy()

# Normalize each metric by its maximum value across both regions
comparison_df_norm["South Asia"] = comparison_df["South Asia"] / comparison_df[["South Asia","Middle East"]].max(axis=1)
comparison_df_norm["Middle East"] = comparison_df["Middle East"] / comparison_df[["South Asia","Middle East"]].max(axis=1)
```

```python
x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, comparison_df_norm["South Asia"], width, label="South Asia", color="skyblue")
plt.bar(x + width/2, comparison_df_norm["Middle East"], width, label="Middle East", color="orange")

plt.xticks(x, comparison_df["Metric"])
plt.ylabel("Normalized Value (0–1)")
plt.title("Comparison of HDI Metrics: South Asia vs Middle East (Normalized by Metric)")
plt.legend()
plt.show()
```

```python
comparison_df["Disparity"] = abs(comparison_df["Middle East"] - comparison_df["South Asia"])
max_disparity_metric = comparison_df.loc[comparison_df["Disparity"].idxmax(), "Metric"]

print("Metric with greatest disparity:", max_disparity_metric)
```

5.HDI Disparity:

• Compute the range (max – min) of HDI for each region.

• Compute the coefficient of variation (CV = std/mean) for HDI.

• Identify which region exhibits more variation in HDI.

```python
import pandas as pd

south_asia_df = pd.read_csv("HDI SouthAsia 2020 2022.csv")
middle_east_df = pd.read_csv("HDI MiddleEast 2020 2022.csv")
```

```python
# South Asia
sa_hdi_range = south_asia_df["hdi"].max() - south_asia_df["hdi"].min()

# Middle East
me_hdi_range = middle_east_df["hdi"].max() - middle_east_df["hdi"].min()

print("HDI Range (South Asia):", sa_hdi_range)
print("HDI Range (Middle East):", me_hdi_range)
```

```python
# South Asia
sa_hdi_cv = south_asia_df["hdi"].std() / south_asia_df["hdi"].mean()

# Middle East
me_hdi_cv = middle_east_df["hdi"].std() / middle_east_df["hdi"].mean()

print("HDI Coefficient of Variation (South Asia):", round(sa_hdi_cv, 3))
print("HDI Coefficient of Variation (Middle East):", round(me_hdi_cv, 3))
```

```python
if sa_hdi_cv > me_hdi_cv:
    more_var_region = "South Asia"
else:
    more_var_region = "Middle East"

print("Region with more HDI variation:", more_var_region)
```

6.Correlation Analysis:

• For each region, compute correlations of HDI with:

– Gender Development Index

– Life Expectancy Index

• Create scatter plots with trendlines for each correlation.

• Interpret the strength and direction of these relationships.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

south_asia_df = pd.read_csv("HDI SouthAsia 2020 2022.csv")
middle_east_df = pd.read_csv("HDI MiddleEast 2020 2022.csv")
```

```python
sa_corr_gender = south_asia_df["hdi"].corr(south_asia_df["gender_development"])
sa_corr_life = south_asia_df["hdi"].corr(south_asia_df["life_expectancy"])

print("South Asia Correlation with Gender Development Index:", round(sa_corr_gender, 3))
print("South Asia Correlation with Life Expectancy Index:", round(sa_corr_life, 3))
```

```python
me_corr_gender = middle_east_df["hdi"].corr(middle_east_df["gender_development"])
me_corr_life = middle_east_df["hdi"].corr(middle_east_df["life_expectancy"])

print("Middle East Correlation with Gender Development Index:", round(me_corr_gender, 3))
print("Middle East Correlation with Life Expectancy Index:", round(me_corr_life, 3))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

metrics = ["gender_development", "life_expectancy"]
regions = {
    "South Asia": south_asia_df,
    "Middle East": middle_east_df
}
colors = {"South Asia": "skyblue", "Middle East": "orange"}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 grid
axes = axes.flatten()  # flatten for easy indexing

plot_idx = 0
for region_name, df in regions.items():
    for metric in metrics:
        sns.regplot(
            x=df[metric],
            y=df["hdi"],
            ci=None,
            ax=axes[plot_idx],
            color=colors[region_name]
        )
        axes[plot_idx].set_title(f"{region_name}: HDI vs {metric.replace('_',' ').title()}")
        axes[plot_idx].set_xlabel(metric.replace('_',' ').title())
        axes[plot_idx].set_ylabel("HDI")
        plot_idx += 1

plt.tight_layout()
plt.show()
```

7.Outlier Detection:

• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.

• Create scatter plots highlighting outliers in a different color.

• Discuss the significance of these outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt

south_asia_df = pd.read_csv("HDI SouthAsia 2020 2022.csv")
middle_east_df = pd.read_csv("HDI MiddleEast 2020 2022.csv")
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

```python
sa_hdi_outliers, sa_hdi_lb, sa_hdi_ub = detect_outliers_iqr(south_asia_df, "hdi")
sa_gni_outliers, sa_gni_lb, sa_gni_ub = detect_outliers_iqr(south_asia_df, "gross_inc_percap")

print("South Asia HDI outliers:")
print(sa_hdi_outliers[["country", "year", "hdi"]])
print("\nSouth Asia GNI per Capita outliers:")
print(sa_gni_outliers[["country", "year", "gross_inc_percap"]])
```

```python
me_hdi_outliers, me_hdi_lb, me_hdi_ub = detect_outliers_iqr(middle_east_df, "hdi")
me_gni_outliers, me_gni_lb, me_gni_ub = detect_outliers_iqr(middle_east_df, "gross_inc_percap")

print("Middle East HDI outliers:")
print(me_hdi_outliers[["country", "year", "hdi"]])
print("\nMiddle East GNI per Capita outliers:")
print(me_gni_outliers[["country", "year", "gross_inc_percap"]])
```

```python
def plot_outliers(df, x_col, y_col, region_name):
    outliers, _, _ = detect_outliers_iqr(df, y_col)

    plt.figure(figsize=(8,5))
    plt.scatter(df[x_col], df[y_col], color="skyblue", label="Normal")
    plt.scatter(outliers[x_col], outliers[y_col], color="red", label="Outliers")
    plt.xlabel(x_col.replace('_',' ').title())
    plt.ylabel(y_col.replace('_',' ').title())
    plt.title(f"{region_name}: {y_col.replace('_',' ').title()} vs {x_col.replace('_',' ').title()} with Outliers")
    plt.legend()
    plt.show()
```

```python
plot_outliers(south_asia_df, "gross_inc_percap", "hdi", "South Asia")
plot_outliers(middle_east_df, "gross_inc_percap", "hdi", "Middle East")
```
